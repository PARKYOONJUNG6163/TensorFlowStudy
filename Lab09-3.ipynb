Wide NN for XOR

```python
import tensorflow as tf
import numpy as np
```

```python
x_data = np.array([[0,0], [0,1], [1,0], [1,1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

W1 = tf.Variable(tf.random_normal([2, 10]), name='weight1')
b1 = tf.Variable(tf.random_normal([10]), name='bias1') # bias는 out의 개수와 같음
layer1 = tf.sigmoid(tf.matmul(X, W1)+b1) #활성함수 시그모이드

W2 = tf.Variable(tf.random_normal([10, 10]), name='weight2')
b2 = tf.Variable(tf.random_normal([10]), name='bias2') # bias는 out의 개수와 같음
layer2 = tf.sigmoid(tf.matmul(layer1, W2)+b2) #활성함수 시그모이드


W3 = tf.Variable(tf.random_normal([10, 10]), name='weight3')
b3 = tf.Variable(tf.random_normal([10]), name='bias3') # bias는 out의 개수와 같음
layer3 = tf.sigmoid(tf.matmul(layer2, W3)+b3) #활성함수 시그모이드

W4 = tf.Variable(tf.random_normal([10, 1]), name='weight4')
b4 = tf.Variable(tf.random_normal([1]), name='bias4') # bias는 out의 개수와 같음
hypothesis = tf.sigmoid(tf.matmul(layer3, W4)+b4) #활성함수 시그모이드

cost = -tf.reduce_sum(Y *tf.log(hypothesis) + (1-Y) * tf.log(1 - hypothesis)) # 축 1로 해서 합을 구하고 평균을 냄
train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

# 4개이므로 test와 train 나누지 않음

predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuarcy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001) :
        _, cost_val = sess.run([train, cost], feed_dict={X: x_data, Y: y_data})
        if step % 100 == 0:
            print(step, cost_val)
    
    h,c,a = sess.run([hypothesis, predicted, accuarcy], feed_dict={X:x_data, Y:y_data})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\nAccuracy: ", a)
    
```
