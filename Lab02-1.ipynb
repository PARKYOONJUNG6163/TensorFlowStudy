```python
import tensorflow as tf
```

```python
# Build graph using TF operations

# H(x) = Wx + b
# 각각 X와 Y data
x_train = [1, 2, 3]
y_train = [1, 2, 3]

# shape을 정의하고 값을 정의하면 됨
# 보통 W와 b의 값을 모르므로 랜덤한 값을 줌 
W = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

# 우리의 가설 Wx + b
hypothesis = x_train * W + b

# reduce_mean 은 어떤 tensor가 주어졌을 때 평균 내주는 것
# t = [1., 2. ,3. , 4.]
# tf.reduce_mean(t) ==> 2.5
cost = tf.reduce_mean(tf.square(hypothesis - y_train))

#GradientDescent
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
# w와 b를 조정해서 스스로 최소화하는 값을 찾음
train = optimizer.minimize(cost)
```

```python
# Run/Update graph and get results
# 그래프구현이 끝났으므로 실행시켜 주기 위해 Session 실행
sess = tf.Session()

# Variable을 실행하기 전에는 반드시 global_variables_initializer()함수를 실행해야 함
sess.run(tf.global_variables_initializer())

# 학습시켜주는 부분
for step in range(2001) :
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```
